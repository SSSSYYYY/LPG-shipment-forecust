```python
import pandas as pd
import urllib2
from urllib import FancyURLopener #create headless browser
from random import choice #this library helps pick a random item from a list
from bs4 import BeautifulSoup
from bs4 import SoupStrainer

user_agents = [
    'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
    'Opera/9.80 (X11; Linux i686; Ubuntu/14.10) Presto/2.12.388 Version/12.16',
    'Mozilla/5.0 (Windows; U; Windows NT 6.1; rv:2.2) Gecko/20110201',
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.75.14 (KHTML, like Gecko) Version/7.0.3 Safari/7046A194A',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'
]

class MyOpener(FancyURLopener,object):
    version = choice(user_agents)
```

```python
urls = []
sec = []

url ='https://www.bloomberg.com/research/sectorandindustry/overview/sectorlanding.asp'
myopener = MyOpener()
page = myopener.open(url)
html = page.read()
soup = BeautifulSoup(html, "lxml")
infotable = soup.find_all('table', {'id':"sectorTable"} )
rows = infotable[0].find_all('h3')
for row in rows:
    a = row.find('a')
    href = a.get('href')
    urls.append('https://www.bloomberg.com/research/sectorandindustry/'+ href[3:])
    sector = str(a.contents[0])
    sec.append(sector)
    

print urls
print sec
```

```python
data = [('sector', sec),
         ('url', urls),
         ]
df = pd.DataFrame.from_items(data)
```
